```python
#   Importowanie bibliotek
from IPython.display import Image
from IPython.core.display import HTML
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
from scipy import interpolate
```

```python
# Podpunkt A
# Deklarowanie macierzy
arr_12_1000 = np.zeros((12,1000))

# Wypełnianie macierzy liczbami losowymi z przedziału (0,1)
for m in range(12):
    for n in range(1000):
        arr_12_1000[m,n] = random.random()


arr_1000_1 = np.zeros((1000,1))

for m in range(1000):
    arr_1000_1[m,0] =  np.sum(arr_12_1000[:,m]) - 6

plt.figure(1)

a = plt.hist(arr_1000_1, bins='auto',density=True)
m = np.shape(a[1])
print(m[0])
print(a[1][0])
print(a[1][m[0]-1])
print(np.shape(a[0]))
x_arr = np.linspace(a[1][0],a[1][m[0]-1],m[0]-1)
# Wygładzanie funkcji gęstości rozkładu
f = interpolate.interp1d(x_arr,a[0],kind='cubic')
# pdf = norm.pdf(a[0])
new_x = np.linspace(a[1][0],a[1][m[0]-1],150)
new_y = f(new_x)
plt.plot(new_x, new_y)

plt.title("Histogram próbki zmiennej losowej stosując regułę tuzina")
plt.legend(['Funkcja gęstości'])
plt.show()
```
